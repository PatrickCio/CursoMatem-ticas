##**Introducción**
Las funciones son relaciones que asocian elementos de un conjunto (dominio) a elementos de otro (codominio). La imagen de un subconjunto del dominio es el conjunto de resultados al aplicar la función a cada elemento del subconjunto, mientras que la preimagen de un subconjunto del codominio es el conjunto de elementos del dominio que se mapean a dicho subconjunto. Estas herramientas son esenciales para analizar el comportamiento de funciones, estudiar transformaciones geométricas y definir nuevos conjuntos en matemáticas.
  
## **Concepto de Función**
Las funciones en matemáticas, sirven para modelar diversas relaciones entre distintos fenómenos o situaciones, que suceden en nuestra vida cotidiana, que tienen una causa y efecto, por ejemplo, la cantidad de kilómetros por hora recorridos por un vehículo depende de la velocidad, el área de un cuadrado depende de la longitud de su lado y el costo de la producción está en función al valor de los materiales utilizados.

# **Introducción a las funciones matemáticas: Imagen y preimagen**

## **Definición de función:**
 Una función se puede representar matemáticamente como una relación entre dos conjuntos, el dominio (D) y el codominio (C), utilizando la notación f: D → C. Esta notación indica que la función f asigna a cada elemento x del dominio D un único elemento y del codominio C. Se puede expresar de otra manera como y = f(x), donde y depende del valor de x.
 Es decir, La primera magnitud es la variable independiente y a la segunda magnitud o imagen (que se deduce de la primera) se dirá que es la variable dependiente. Por ejemplo, si la variable independiente es x, la variable dependiente será f (x), que se lee “f de x”, la cual generalmente se designa con la letra y. Entonces, se dirá que y es función de x, o que y depende de x.

# **Propiedades de las funciones:**
1. Unicidad: A cada elemento del dominio le corresponde un único elemento del codominio. Matemáticamente, si x1 y x2 son elementos del dominio y f(x1) = f(x2), entonces x1 = x2.
2. Existencia: No todos los elementos del codominio tienen que ser la imagen de algún elemento del dominio. Es decir, la imagen de la función puede no ser todo el codominio.

# **Tipos de funciones:**
1. Función inyectiva (o uno a uno): Si para cada elemento y del codominio hay a lo sumo un elemento x del dominio tal que f(x) = y. Matemáticamente, si x1 y x2 son elementos del dominio y f(x1) = f(x2), entonces x1 = x2.
2. Función suryectiva (o sobreyectiva): Si la imagen de la función es todo el codominio. Matemáticamente, para todo elemento y del codominio, existe al menos un elemento x del dominio tal que f(x) = y.
3. Función biyectiva (o uno a uno y sobreyectiva): Si es tanto inyectiva como suryectiva. Es decir, si para cada elemento y del codominio hay un único elemento x del dominio tal que f(x) = y, y la imagen de la función es todo el codominio.

# **Ejemplos de funciones:**
1. Función lineal: f(x) = mx + b, donde m y b son constantes y m ≠ 0.
- Es inyectiva si m ≠ 0.
- Es sobreyectiva para cualquier valor de m y b.
- Es biyectiva si m ≠ 0.
2. Función cuadrática: f(x) = ax² + bx + c, donde a, b y c son constantes y a ≠ 0.
- No es inyectiva en general.
- Es sobreyectiva si a > 0.
- No es biyectiva en general.
3. Función exponencial: f(x) = a^x, donde a es una constante positiva y a ≠ 1.
- Es inyectiva para cualquier valor de a > 0.
- Es suryectiva para cualquier valor de a > 0.
- Es biyectiva para cualquier valor de a > 0.
4. Función logarítmica: f(x) = log_a(x), donde a es una constante positiva y a ≠ 1.
- Es la función inversa de la función exponencial.
- Es inyectiva para x > 0.
- Es suryectiva para x > 0.
- Es biyectiva para x > 0.

# **Imagen y Preimagen:**
Una función es una relación entre dos magnitudes o cantidades, por ejemplo x y f(x), de manera que a cada valor de la primera magnitud llamada preimagen, le corresponde un único valor de la segunda, llamada imagen.

- Imagen: La imagen de un conjunto A bajo una función f se denota como f(A) y se define como el conjunto de todos los elementos y del codominio C que son la imagen de algún elemento x del dominio D que pertenece a A. Matemáticamente, f(A) = {y ∈ C | ∃x ∈ A, f(x) = y}.
- Preimagen: La preimagen de un conjunto B bajo una función f se denota como f⁻¹(B) y se define como el conjunto de todos los elementos x del dominio D que se asocian a algún elemento y del codominio C que pertenece a B. Matemáticamente, f⁻¹(B) = {x ∈ D | f(x) ∈ B}.

# Cálculo de Preimágenes

## Función Lineal: \( f(x) = mx + b \)

#### Ejemplo
Supongamos que tenemos la función \( f(x) = 2x + 3 \) y queremos encontrar la preimagen del conjunto \( B = \{7, 11\} \).

##### Cálculo
Para \( y = 7 \):
\[ 7 = 2x + 3 \]
\[ 2x = 4 \]
\[ x = 2 \]

Para \( y = 11 \):
\[ 11 = 2x + 3 \]
\[ 2x = 8 \]
\[ x = 4 \]

###### Resultado
La preimagen de \( B \) bajo la función \( f(x) = 2x + 3 \) es:
\[ f^{-1}(B) = \{2, 4\} \]

## Función Cuadrática: \( f(x) = ax^2 + bx + c \)

#### Ejemplo
Supongamos que tenemos la función \( f(x) = x^2 - 4 \) y queremos encontrar la preimagen del conjunto \( B = \{0\} \).

##### Cálculo
Para \( y = 0 \):
\[ 0 = x^2 - 4 \]
\[ x^2 = 4 \]
\[ x = \pm 2 \]

### Resultado
La preimagen de \( B \) bajo la función \( f(x) = x^2 - 4 \) es:
\[ f^{-1}(B) = \{-2, 2\} \]

## Función Exponencial: \( f(x) = a^x \)

#### Ejemplo
Supongamos que tenemos la función \( f(x) = 2^x \) y queremos encontrar la preimagen del conjunto \( B = \{8, 16\} \).

##### Cálculo
Para \( y = 8 \):
\[ 8 = 2^x \]
\[ x = \log_2(8) \]
\[ x = 3 \]

Para \( y = 16 \):
\[ 16 = 2^x \]
\[ x = \log_2(16) \]
\[ x = 4 \]

### Resultado
La preimagen de \( B \) bajo la función \( f(x) = 2^x \) es:
\[ f^{-1}(B) = \{3, 4\} \]

## Función Logarítmica: \( f(x) = \log_a(x) \)

#### Ejemplo
Supongamos que tenemos la función \( f(x) = \log_2(x) \) y queremos encontrar la preimagen del conjunto \( B = \{1, 2\} \).

##### Cálculo
Para \( y = 1 \):
\[ 1 = \log_2(x) \]
\[ x = 2^1 \]
\[ x = 2 \]

Para \( y = 2 \):
\[ 2 = \log_2(x) \]
\[ x = 2^2 \]
\[ x = 4 \]

### Resultado
La preimagen de \( B \) bajo la función \( f(x) = \log_2(x) \) es:
\[ f^{-1}(B) = \{2, 4\} \]

## Definición Precisa de Función
Una función necesita más que solo una regla de asociación; requiere:
  - **Un nombre**.
  - **Una regla de asociación**.
  - **Un espacio de partida** (dominio).
  - **Un espacio de llegada** (codominio).

Una función se define formalmente como una regla de asociación que asigna a cada elemento \\( x \\) de un conjunto \\( A \\) un único elemento \\( y \\) de un conjunto \\( B \\), denotado como \\( f(x) \\).
  
## **Elementos de una Función**
  - **Nombre de la función**: \\( f \\).
  - **Regla de asociación**: \\( f(x) \\).
  - **Dominio**: Espacio del que se parte (por ejemplo, números reales ℝ)\.
  - **Codominio**: Espacio al que se llega (por ejemplo, números reales ℝ) \.
  - **Imagen**: El único elemento \\( y \\) en \\( B \\) asociado a cada \\( x \\) en \\( A \\).

  Nota: Existen funciones numéricas pero también pueden ser no numéricas. Por ejemplo:  𝐴= {palabras del castellano},  𝐵= {a,b,...z}, entonces podemos crear la función  𝐼:𝐴−>𝐵 ,  𝑥 -> 𝑓(𝑥)= letra inicial de cada palabra, entonces  𝐼 (marsupial)=m

## **Expresión Matemática de una Función**
Una función \\( f \\) que parte de un conjunto \\( A \\) y llega a un conjunto \\( B \\) se denota \\( f: A \\ \\).

## **Completar la Noción de Función**
Es necesario definir todos los elementos de la función, no solo la regla de asociación. Por ejemplo, \\( f(x) = 3x^2 \\) necesita especificar que \\( x \\) pertenece a los números reales ℝ para ser una función completa. Así, \\( f \\) puede ser una función desde los ℝ→ℝ.
  
## **Importancia de la Unicidad**
Es crucial que la regla de asociación asigne un único elemento a cada \\( x \\) en el dominio. Esto es una característica esencial de las funciones."

##**Importancia de la Existencia**
La existencia en una función se refiere a la posibilidad de calcular un valor de salida para un determinado valor de entrada. Es fundamental para analizar el comportamiento de la función, definir propiedades como la continuidad y realizar operaciones matemáticas como límites, derivadas e integrales. La existencia o no de una función en un punto específico determina su comportamiento y sus aplicaciones.
