## 1. Funciones de utilidad en ciencias sociales:

### Definición:

La función de utilidad, representada como $( U(x, y) )$, representa la satisfacción o beneficio que un individuo obtiene al consumir bienes o servicios. Las variables $( x ) e ( y )$ representan las cantidades de dos bienes específicos. Se busca encontrar la combinación de bienes que maximice la utilidad del individuo, considerando su presupuesto y preferencias.
##Aplicación de la función de utilidad en ciencia sociales
Los economistas conciben las desiciones de los individuos es a través de una función de utilidad, en el cual es un tipo de función en dónde el espacio de partida es un espacio de bienes o tambien es una función que muestra la satisfacción del individuo con el consumo.

### Propiedades:

- **Monotonicidad**: La utilidad aumenta a medida que aumenta la cantidad de cada bien. Esto significa que el individuo obtiene mayor satisfacción al consumir más de ambos bienes.
- **Tasa marginal decreciente de sustitución**: La cantidad de un bien que un individuo está dispuesto a renunciar para obtener una unidad más del otro bien disminuye a medida que se consume más de este último. En otras palabras, la satisfacción adicional que se obtiene por una unidad más de un bien disminuye a medida que se consume más de ese bien.

### Aplicaciones:
- **Teoría de la Elección del Consumidor:** Esta teoría utiliza la función de utilidad para modelar cómo los consumidores eligen entre diferentes combinaciones de bienes y servicios dentro de sus restricciones presupuestarias. Por ejemplo, mientras más agua se dispone mas utilidad tenemos.

**Ejemplo:** Supongamos que un consumidor tiene una utilidad $( U(x, y) = (x \ y )$ donde \( x \) es la cantidad de pan y \( y \) la cantidad de agua. El consumidor elige la combinación de \( x \) e \( y \) que maximiza su utilidad dado su presupuesto.

**Forma de la función de utilidad:**

Usualmente representada con una pendiente positiva que disminuye a medida que se consume más. Ejemplos: funciones como $( log(x) )$ y $( x^x )$.

 Es creciente y cóncava, indicando que el individuo prefiere más de un bien a menos del mismo y una combinación de dos bienes a una cantidad mayor de un solo bien


Nota: Cuando tenemos dos bienes pasamos a representar la función de utilidad en un eje tridimencional y se representa en un eje tridimensional: un bien en el eje 𝑥 , otro en el eje 𝑦, y la utilidad en el eje 𝑧.

- **Análisis del comportamiento del consumidor**: Las funciones de utilidad se utilizan para modelar las decisiones de los consumidores sobre cómo asignar sus recursos para maximizar su satisfacción. Se pueden analizar diferentes escenarios, como el efecto del precio, la calidad y la disponibilidad de los bienes en las decisiones de compra.

- **Teoría del bienestar**: Las funciones de utilidad se emplean para evaluar el bienestar social y las políticas públicas que afectan el consumo de bienes y servicios. Se pueden comparar los niveles de bienestar de diferentes individuos o grupos sociales, y analizar el impacto de políticas como impuestos, subsidios o programas de bienestar social.

### Ejemplo:

Un estudiante tiene un presupuesto de $100 para gastar en libros y música. Su función de utilidad es U(libros, música) = 2 libros + 3 música. ¿Cuántas unidades de libros y música debe comprar el estudiante para maximizar su utilidad?

Para resolver este problema, se puede plantear un problema de optimización matemática que maximice la función de utilidad U(libros, música) sujeta a la restricción presupuestaria del estudiante. La solución a este problema indicará la cantidad óptima de libros y música que debe comprar el estudiante para maximizar su satisfacción.

## Elección Intertemporal:

**Consumo y Ahorro:** La función de utilidad ayuda a modelar decisiones intertemporales, como el consumo y el ahorro.

**Ejemplo:** Un consumidor puede tener una función de utilidad que depende del consumo presente y futuro $( U(C_1, C_2) = C_1 \cdot C_2 )$, donde $( C_1 )$ es el consumo en el periodo 1 y $( C_2 )$ el consumo en el periodo 2. El consumidor decide cuánto consumir y ahorrar hoy para maximizar su utilidad total.

En conclusión, Las funciones de utilidad son una herramienta poderosa para analizar el comportamiento económico. Son un concepto fundamental en economía y se utilizan en una amplia gama de aplicaciones.

## 2. Funciones sobreyectivas, inyectivas y biyectivas:

### Función sobreyectiva:

Una función \( f: A→B \) es sobreyectiva si la imagen de \( f \) cubre todo el codominio \( B \). En otras palabras, para cada elemento \( y \) en \( B \), existe al menos un elemento \( x \) en \( A \) tal que \( f(x) = y \). Esto significa que todos los elementos del codominio \( B \) tienen al menos una preimagen en el dominio \( A \).

Una función es sobreyectiva cuando todos los elementos del codominio son imagen de al menos un elemento del dominio, es decir, el codominio es igual al recorrido.
En lenguaje algebraico, una función f de A en B es sobreyectiva cuando:

$Rec (f) = B$

### Función inyectiva:

Una función \( f: A→B \) es inyectiva si cada elemento del dominio se asocia a un único elemento del codominio. Es decir, si \( f(x1) = f(x2) \), entonces \( x1 = x2 \). Esto significa que no hay elementos repetidos en la imagen de la función. Es decir, si una función es inyectiva cuando cada elemento del recorrido es imagen de sólo un elemento del dominio, es decir, ningún elemento del recorrido es imagen de dos preimágenes diferentes.

En lenguaje algebraico, una función f de A en B es inyectiva si x e y pertenecen al conjunto A (dominio) y, necesariamente se cumple que:
- f (x) = f (y) x = y

Dicho de otra manera:
- f (x) ≠ f (y) x ≠ y

### Función biyectiva:

Una función $( f: A → B )$ es biyectiva si es tanto inyectiva como sobreyectiva. Es decir, existe una función inversa $( f⁻¹: B → A )$ que cumple que $( f⁻¹(f(x)) = x )$ para todo $( x )$ en $( A )$ y $( f(f⁻¹(y)) = y )$ para todo $( y )$ en $( B )$. Esto significa que la función tiene una correspondencia uno a uno entre los elementos del dominio y del codominio, y existe una función inversa que deshace la acción de la función original.

En lenguaje algebraico, una función f de A en B es biyectiva si se cumple que, siendo x e y
elementos de A:

f (x) = f (y) x = y

Rec (f) = B

### Importancia y Aplicaciones:

- **Matemáticas:** Estas propiedades son fundamentales en álgebra, análisis y topología.
- **Informática:** En la programación, las funciones biyectivas aseguran que cada entrada tiene una única salida y viceversa, lo cual es útil en algoritmos de encriptación.
- **Economía:** Las funciones inyectivas pueden modelar situaciones donde no hay duplicación de recursos, y las sobreyectivas aseguran que todos los recursos son utilizados.
- **Teoría de conjuntos**: Las funciones sobreyectivas, inyectivas y biyectivas se utilizan para establecer relaciones entre conjuntos y para estudiar su estructura. Se pueden clasificar conjuntos según las propiedades de las funciones que los relacionan.
- **Criptografía**: Las funciones biyectivas se emplean para diseñar algoritmos de cifrado y descifrado seguros. Se utilizan funciones que son difíciles de invertir para proteger información confidencial.

## 3. Funciones compuestas:

### Definición:

La composición de dos funciones \( f \) y \( g \) se define como la función \( h(x) = g(f(x)) \), donde \( x \) pertenece al dominio de \( f \). La función \( h \) representa la acción de aplicar primero la función \( f \) y luego la función \( g \) a un valor de entrada.

### Propiedades:

- **Asociatividad:** La composición de funciones es asociativa: (h ° g) ° f = h ° (g ° f). Esto significa que el orden en que se agrupan las funciones compuestas no afecta el resultado final.
- **Conmutatividad:** La composición de funciones es conmutativa solo en algunos casos específicos. En general, el orden en que se aplican las funciones puede afectar el resultado.
-**No necesariamente conmutativa:** $( f \circ g \neq g \circ f )$ en general.


### Aplicaciones:

- **Matemáticas:** Composición de funciones en cálculos derivados e integrales.
- **Ciencias de la Computación:** En programación funcional, las funciones compuestas permiten encadenar operaciones de manera eficiente.
- **Economía:** Modelar procesos económicos donde varias funciones dependientes están encadenadas.
- **Cálculo**: Las funciones compuestas se emplean para derivar nuevas funciones a partir de funciones básicas. Se pueden utilizar reglas de diferenciación para calcular la derivada de una función compuesta.

### Ejemplo:

- Función compuesta para calcular el costo total de un viaje: Supongamos que la función \( f(x) \) representa la distancia recorrida en un viaje en función del tiempo \( x \), y la función \( g(x) \) representa el costo del viaje en función de la distancia \( x \). La función compuesta \( h(x) = g(f(x)) \) representaría el costo total del viaje en función del tiempo.

## 4. Función inversa:

### Definición:

La función inversa de una función biyectiva f: A → B se define como la función f⁻¹: B → A que cumple que f⁻¹(f(x)) = x para todo x en A y f(f⁻¹(y)) = y para todo y en B. La función inversa deshace la acción de la función original, permitiendo encontrar la preimagen de un elemento en el codominio.


### Propiedades:

- **Existencia**: La función inversa solo existe para funciones biyectivas. No todas las funciones tienen una función inversa.
- **Unicidad**: La función inversa es única para cada función biyectiva. Existe una sola función inversa que deshace la acción de la función original.
- **Simetría**: La gráfica de una función biyectiva y su función inversa son simétricas respecto a la diagonal \( y = x \). Esto significa que si se intercambian los ejes \( x \) e \( y \) en la gráfica de la función original, se obtiene la gráfica de la función inversa.

### Aplicaciones:

- Resolución de ecuaciones: La función inversa se utiliza para encontrar el valor de x que satisface una ecuación de la forma f(x) = y. La solución a la ecuación se obtiene aplicando la función inversa a ambos lados de la ecuación: x = f⁻¹(y).
- Criptografía: Las funciones inversas se emplean para diseñar algoritmos de cifrado y descifrado seguros. Se utilizan funciones que son difíciles de invertir para proteger información confidencial.


### Ejemplo:

Función inversa para calcular el tiempo de viaje: Supongamos que la función f(x) representa la distancia recorrida en un viaje en función del tiempo x, como en el ejemplo anterior. La función inversa f⁻¹(x) representaría el tiempo necesario para recorrer una distancia específica x. Si conocemos la distancia que queremos recorrer, podemos usar la función inversa para calcular el tiempo que tomará el viaje.
