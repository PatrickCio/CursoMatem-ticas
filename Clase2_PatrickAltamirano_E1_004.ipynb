## Introducción
####Los algoritmos y sistemas numéricos son fundamentales en la computación y las matemáticas. Los algoritmos son secuencias de pasos bien definidos para resolver problemas o realizar tareas, mientras que los sistemas numéricos son métodos para representar y manipular números. Juntos, permiten la resolución eficiente de problemas complejos en áreas como la informática, la ingeniería y las ciencias, abarcando desde la búsqueda y el ordenamiento de datos hasta la conversión y manipulación de diferentes sistemas numéricos.



##1. Algoritmos
####Un algoritmo de programación son un conjunto ordenado de pasos finito, con el fin de resolver un problema o realizar una tarea.

##-Tipos de Algoritmos de Programación

Existen varios tipos de algoritmos de programación, cada uno diseñado para resolver diferentes tipos de problemas, los algoritmos más conocidos son:

1. **Algoritmos de Búsqueda**: Encuentran un elemento específico en una estructura de datos (e.g., búsqueda lineal, búsqueda binaria).
2. **Algoritmos de Ordenamiento**: Organizan elementos en una secuencia o arreglo (e.g., ordenamiento burbuja, quicksort, mergesort).
3. **Algoritmos Recursivos**: Resuelven problemas dividiéndolos en subproblemas más pequeños (e.g., factorial, Fibonacci).
4. **Algoritmos de Divide y Vencerás**: Dividen un problema en partes más manejables y combinan los resultados (e.g., mergesort, quicksort).
5. **Algoritmos Dinámicos**: Resuelven problemas complejos almacenando resultados intermedios para evitar cálculos redundantes (e.g., Fibonacci, programación dinámica).
6. **Algoritmos Greedy (Ávidos)**: Toman decisiones óptimas locales con la esperanza de encontrar una solución global óptima (e.g., problema del cambio de monedas, algoritmo de Kruskal).
7. **Algoritmos de Vuelta Atrás (Backtracking)**: Buscan todas las posibles soluciones probando y descartando soluciones parciales (e.g., sudoku, n-reinas).
8. **Algoritmos de Grafos**: Trabajan con estructuras de grafos para encontrar caminos, ciclos, y conexiones (e.g., algoritmo de Dijkstra, algoritmo de Floyd-Warshall).
9. **Algoritmos de Transformación de Cadenas**: Procesan y transforman cadenas de caracteres (e.g., algoritmos de búsqueda de patrones, compresión de datos).

Nota: Estos algoritmos se utilizan en una variedad de campos, desde la informática y la ingeniería hasta las ciencias y las matemáticas, para resolver problemas de manera eficiente y efectiva.

## Algoritmos de Búsqueda: Búsqueda Lineal y Búsqueda Binaria

### Búsqueda Lineal

La búsqueda lineal es un método simple para encontrar un elemento en una lista no ordenada. Se revisa cada elemento de la lista uno por uno, desde el inicio hasta el final, hasta encontrar el elemento objetivo o llegar al final de la lista.

#### Ejemplo de Búsqueda Lineal

Supongamos que tenemos una lista `[3, 8, 2, 7, 5]` y queremos encontrar el número `7`.

1. Comenzamos en el primer elemento: `3` (no es `7`).
2. Pasamos al siguiente elemento: `8` (no es `7`).
3. Luego: `2` (no es `7`).
4. Después: `7` (es el objetivo, búsqueda completa).

En el peor de los casos, debemos revisar todos los elementos, resultando en una complejidad de tiempo de O(n).

### Búsqueda Binaria

La búsqueda binaria es más eficiente, pero requiere que la lista esté previamente ordenada. Divide repetidamente el rango de búsqueda a la mitad, comparando el elemento objetivo con el elemento medio de la lista.

#### Ejemplo de Búsqueda Binaria

Supongamos que tenemos una lista ordenada `[2, 3, 5, 7, 8]` y queremos encontrar el número `7`.

1. Comenzamos en el elemento medio: `5` (no es `7`, pero `7` es mayor que `5`).
2. Buscamos en la mitad derecha: `[7, 8]`.
3. El elemento medio de la nueva sublista es `7` (es el objetivo, búsqueda completa).

La búsqueda binaria tiene una complejidad de tiempo de O(log n) debido a la reducción del rango de búsqueda a la mitad en cada paso.

###¿Cuál es más eficiente para encontrar datos?

Para listas largas, la búsqueda binaria es significativamente más eficiente que la búsqueda lineal debido a su menor complejidad de tiempo. Mientras que la búsqueda lineal puede requerir revisar cada elemento, la búsqueda binaria reduce el número de comparaciones necesarias dividiendo el rango de búsqueda.

##2. Sistemas Numéircos


Un sistema numérico tiene como objetivo permitir el conteo de los elementos de un conjunto. Se conforma por \( n \) unidades en orden sucesivo que aumentan de \( n \) en \( n \). De acuerdo a \( n \), se define el número de unidades necesarias para pasar de un orden a otro.

Una de las condiciones para utilizar algún sistema numérico es que permita realizar operaciones básicas sobre el conjunto de una forma sencilla.

Otra condición es que por cada elemento debe corresponderle un símbolo escrito.

Cabe mencionar que un elemento en el conjunto  siempre podrá tener dos clases de valores:

- **El valor individual**, por ejemplo, en 58, el 8 representa 8 unidades en el sistema decimal.
- **El valor que le corresponde de acuerdo a su posición**, por ejemplo, en 80, el 8 representa 8 decenas.

##Tipos de sistemas numéicos

####Los sistemas numéricos son métodos para representar números utilizando símbolos y reglas específicas. Aquí hay una breve descripción de algunos sistemas numéricos comunes:

####-Sistema Decimal (Base 10):

Utiliza diez símbolos: 0, 1, 2, ..., 9.
Cada posición en un número decimal representa potencias de diez.

####-Sistema Binario (Base 2):

Utiliza dos símbolos: 0 y 1.
Cada posición en un número binario representa potencias de dos.
Ampliamente utilizado en sistemas digitales y computación.

####-Sistema Octal (Base 8):

Utiliza ocho símbolos: 0, 1, 2, ..., 7.
Cada posición representa potencias de ocho.
Menos común en computación moderna, pero usado en sistemas antiguos y en algunos contextos específicos.

####-Sistema Hexadecimal (Base 16):

Utiliza dieciséis símbolos: 0-9 y A-F (donde A=10, B=11, ..., F=15).
Cada posición representa potencias de dieciséis.
Muy útil para representar números binarios de manera compacta y en programación informática.

##Conclusiones
En conclusión, los algoritmos de programación son herramientas fundamentales en informática que permiten resolver problemas de manera estructurada y eficiente. Se diseñan utilizando una combinación de lógica, matemáticas y razonamiento abstracto para transformar datos de entrada en resultados deseados. La eficiencia de un algoritmo se evalúa en términos de tiempo y espacio, buscando minimizar el tiempo de ejecución y el uso de recursos computacionales. Además, los algoritmos pueden clasificarse según su complejidad y aplicabilidad, desde simples como la búsqueda lineal hasta complejos como los algoritmos de aprendizaje automático. Es por eso que es muy dominar los algoritmos es crucial para desarrollar software robusto, escalable y eficiente.

Cada sistema numérico, como el decimal, binario, octal y hexadecimal, utiliza un conjunto específico de símbolos y reglas para representar números. Estos sistemas no solo permiten contar y realizar operaciones básicas de manera eficiente, sino que también facilitan la comunicación y el almacenamiento de datos en diversos dispositivos y aplicaciones.

En resumen, los sistemas numéricos son un componente esencial en la representación y manipulación de cantidades, jugando un papel crucial en campos tan diversos como la informática, las matemáticas y la ingeniería.
