#1. Lógica Proposicional

## Introducción
#### La lógica proposicional es la rama de la lógica que se ocupa de las proposiciones y sus conectores lógicos.
#### Las proposiciones son afirmaciones que pueden ser verdaderas o falsas. No obstante, no puede ser ambas.

##Conectores lógicos
####∧: Conjunción ("y")
####∨: Disyunción ("o").
####~: Negación ("No").
####⇒: Implicancia ("entonces").
####⇔: Equivalencia ("si y sólo si").

##¿Cómo funcionan los conectores lógicos?

### Negación ("No")
#### La negación de una proposición p es verdadera si p es falso, y falso si p es verdadero.


### Conjunción ("y")
#### La conjunción de p y q es verdadero solo si ambas son verdaderas.

### Disyunción ("o")
#### La disyunción de p y q es verdadero si al menos una es verdadera.


### Implicancia ("entonces")
#### La implicación p -> q es falso solo si p es verdadero y q es falso.


### Equivalencia o Bicondicional ("si y solo si)
#### La equivalencia p <-> q es verdadero si p y q tienen el mismo valor de verdad.

##Tablas de verdad
###Ejemplo de tabla de verdad para P→ Q:


####|P|Q|P → Q|
####|V|V|	 V |
####|V|F|	 F |
####|F|V|	 V |
####|F|F|	 V |

Nota: La implicación solo es falsa cuando el antecedente es verdadero y el consecuente es falso.

##Cuantificadores lógicos
####Existencial (∃): "Existe"
####Universal (∀): "Para todo"
###Ejemplos:

####∃x ∀y: "Existe un x para todo y"
####∀y ∃x: "Para todo y existe un x"

Importante: El orden de los cuantificadores afecta el significado lógico.

###Analogía: "Todas las personas tienen un cumpleaños, ninguno como el mío."

####∃x∀y: Existe un cumpleaños para todas las personas (incorrecto)

####∀y∃x: Para toda persona existe un cumpleaños (correcto)


##Negación de cuantificadores
####Negación de ∃x P(x): ∀x ¬P(x)
####Negación de ∀x P(x): ∃x ¬P(x)

#2. Conjuntos
### Un conjunto es una colección de elementos o objetos, los cuales pueden ser: matemáticos, formas geométricas, etc. Se denota típicamente usando llaves: por ejemplo,     A = {1, 2, 3}.

#### Para denotar un conjunto se utilizara letras mayúsculas como: A, B, C.
#### Para denotar elementos se utiliza letras minúsculas como: a, b, c.

###Notación
#####Un conjunto se denota con llaves: {a, b, c}
#####Conjunto vacío: {}

###Operaciones con Conjuntos
#####Unión (A ∪ B)
#####Intersección (A ∩ B)
#####Diferencia (A - B)
#####Complemento (A^c)

### Explicación de las Propiedades de los Conjuntos

#### Subconjunto: Un conjunto A es subconjunto de B si todos los elementos de A están en B. Se denota A ⊆ B.
#### Conjunto Universal: El conjunto que contiene todos los elementos bajo consideración.
#### Conjunto Vacío: El conjunto que no contiene elementos, denotado ∅.

##Conclusión
####Se puede concluir que estos temas sobre la lógica proposicional y la teoría de conjuntos son fundamentales en matemáticas y computación. La lógica proposicional permite formalizar y evaluar argumentos mediante proposiciones y conectores lógicos, mientras que la teoría de conjuntos proporciona un marco para organizar y manipular colecciones de elementos a través de operaciones como la unión y la intersección. Estos conceptos son esenciales para resolver problemas complejos y estructurar datos de manera lógica y coherente en diversas aplicaciones científicas y tecnológicas.
